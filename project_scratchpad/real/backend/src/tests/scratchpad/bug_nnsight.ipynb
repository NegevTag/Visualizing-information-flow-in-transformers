```python
import nnsight

from info_flow.config import Config
def get_model(model_name: str, hf_token: str) -> nnsight.LanguageModel:
    model_kwargs_dict = {"token": hf_token}
    return nnsight.LanguageModel(model_name, **model_kwargs_dict)  # type: ignore[arg-type]

config = Config()
model = get_model(model_name="meta-llama/Llama-3.1-8B",hf_token=config.hf_token)
with model.trace("this is a test",remote = True):
    print("a")
```
